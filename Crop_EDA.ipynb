```python
# Growth Analysis: Investigate if certain crops or regions show increasing or decreasing trends in yield or production.

# ✨ Set improved aesthetic
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'axes.titlesize': 16,
    'axes.labelsize': 13,
    'axes.titleweight': 'bold',
    'legend.fontsize': 11,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11
})

# ✅ Data cleaning
crop_df_filtered['value'] = pd.to_numeric(crop_df_filtered['value'], errors='coerce')
crop_df_filtered['year'] = pd.to_numeric(crop_df_filtered['year'], errors='coerce')

# ✅ Yield growth by crop
yield_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'yield'].copy()
crop_yield = (
    yield_df.groupby(['item', 'year'])['value']
    .mean()
    .reset_index()
    .sort_values(['item', 'year'])
)
crop_yield['Yield Growth (%)'] = crop_yield.groupby('item')['value'].pct_change() * 100

# Top 6 crops
top_crops = crop_yield['item'].value_counts().head(6).index.tolist()
crop_plot_data = crop_yield[crop_yield['item'].isin(top_crops)]

# ✅ Production growth by region
prod_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'production'].copy()
region_prod = (
    prod_df.groupby(['area', 'year'])['value']
    .sum()
    .reset_index()
    .sort_values(['area', 'year'])
)
region_prod['Production Growth (%)'] = region_prod.groupby('area')['value'].pct_change() * 100

# Top 6 regions
top_regions = region_prod['area'].value_counts().head(6).index.tolist()
region_plot_data = region_prod[region_prod['area'].isin(top_regions)]

# 📊 Combined plots
fig, axes = plt.subplots(1, 2, figsize=(18, 7), sharex=True)

# 🎨 Custom color palettes
crop_palette = sns.color_palette("Set2", len(top_crops))
region_palette = sns.color_palette("Dark2", len(top_regions))

# ▶️ Yield Growth Plot
sns.lineplot(
    data=crop_plot_data,
    x='year', y='Yield Growth (%)', hue='item',
    marker='o', linewidth=2.5, markersize=7, ax=axes[0], palette=crop_palette
)
axes[0].axhline(0, color='gray', linestyle='--', linewidth=1)
axes[0].set_title("Yield Growth by Crop")
axes[0].set_ylabel("Year-on-Year Growth (%)")
axes[0].set_xlabel("Year")
axes[0].legend(title="Crop", bbox_to_anchor=(0.5, -0.35), loc='lower center',ncol=3, borderaxespad=0)

# ▶️ Production Growth Plot
sns.lineplot(
    data=region_plot_data,
    x='year', y='Production Growth (%)', hue='area',
    marker='o', linewidth=2.5, markersize=7, ax=axes[1], palette=region_palette
)
axes[1].axhline(0, color='gray', linestyle='--', linewidth=1)
axes[1].set_title("Production Growth by Region")
axes[1].set_ylabel("Year-on-Year Growth (%)")
axes[1].set_xlabel("Year")
axes[1].legend(title="Region", bbox_to_anchor=(0.5, -0.35), loc='lower center',ncol=4, borderaxespad=0)

# 🌟 Main title
plt.suptitle("Growth Trends in Yield and Production", fontsize=18, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

```python
# Input-Output Relationships
# Study correlations between Area harvested, Yield, and Production to understand the relationship between land usage and productivity.

# ✨ Style settings
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'axes.titlesize': 16,
    'axes.labelsize': 13,
    'axes.titleweight': 'bold',
    'legend.fontsize': 11,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11
})

# ✅ Data cleaning
crop_df_filtered['value'] = pd.to_numeric(crop_df_filtered['value'], errors='coerce')
crop_df_filtered['year'] = pd.to_numeric(crop_df_filtered['year'], errors='coerce')

# ✅ Separate dataframes
area_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'area harvested'].copy()
area_df = area_df.rename(columns={'value': 'Area harvested (ha)'})

yield_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'yield'].copy()
yield_df = yield_df.rename(columns={'value': 'Yield (kg/ha)'})

prod_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'production'].copy()
prod_df = prod_df.rename(columns={'value': 'Production (tons)'})

# ✅ Merge on item, area, year
merged_df = area_df[['item', 'area', 'year', 'Area harvested (ha)']].merge(
    yield_df[['item', 'area', 'year', 'Yield (kg/ha)']],
    on=['item', 'area', 'year'],
    how='inner'
).merge(
    prod_df[['item', 'area', 'year', 'Production (tons)']],
    on=['item', 'area', 'year'],
    how='inner'
)

# ✅ Correlation matrix
corr_matrix = merged_df[['Area harvested (ha)', 'Yield (kg/ha)', 'Production (tons)']].corr()
print("📊 Correlation Matrix:\n", corr_matrix)

# ✅ Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='YlGnBu',
    linewidths=0.5,
    fmt=".2f",
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Heatmap: Area, Yield & Production", fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# Calculate productivity ratios: Production/Area harvested to cross-verify yields.

# ✨ Style
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'axes.titlesize': 16,
    'axes.labelsize': 13,
    'axes.titleweight': 'bold',
    'legend.fontsize': 11,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11
})

# ✅ Data cleaning
crop_df_filtered['value'] = pd.to_numeric(crop_df_filtered['value'], errors='coerce')
crop_df_filtered['year'] = pd.to_numeric(crop_df_filtered['year'], errors='coerce')

# ✅ Separate Area harvested and Production
area_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'area harvested'].copy()
area_df = area_df.rename(columns={'value': 'Area harvested (ha)'})

prod_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'production'].copy()
prod_df = prod_df.rename(columns={'value': 'Production (tons)'})

# ✅ Merge on item, area, year
merged_df = pd.merge(
    area_df[['item', 'area', 'year', 'Area harvested (ha)']],
    prod_df[['item', 'area', 'year', 'Production (tons)']],
    on=['item', 'area', 'year'],
    how='inner'
)

# ✅ Calculate productivity ratio
merged_df['Productivity (kg/ha)'] = (merged_df['Production (tons)'] * 1000) / merged_df['Area harvested (ha)']

# ✅ Also get reported yield to cross-check
yield_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'yield'].copy()
yield_df = yield_df.rename(columns={'value': 'Reported Yield (kg/ha)'})

# ✅ Merge with reported yield
final_df = pd.merge(
    merged_df,
    yield_df[['item', 'area', 'year', 'Reported Yield (kg/ha)']],
    on=['item', 'area', 'year'],
    how='inner'
)

# ✅ Compare: Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=final_df,
    x='Reported Yield (kg/ha)',
    y='Productivity (kg/ha)',
    hue='item',
    alpha=0.7,
    s=60
)
plt.plot([0, final_df[['Reported Yield (kg/ha)', 'Productivity (kg/ha)']].max().max()],
         [0, final_df[['Reported Yield (kg/ha)', 'Productivity (kg/ha)']].max().max()],
         color='red', linestyle='--', linewidth=1)

plt.title("Reported Yield vs Calculated Productivity")
plt.xlabel("Reported Yield (kg/ha)")
plt.ylabel("Calculated Productivity (kg/ha)")
plt.legend(title="Crop", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# ✅ Print few rows to verify numerically
print(final_df[['item', 'area', 'year', 'Reported Yield (kg/ha)', 'Productivity (kg/ha)']].head(10))
```

```python
# Outliers and Anomalies
''' Identify anomalies in Yield or Production, such as unusually high or low values, and correlate them 
with potential external factors like policies or environmental changes'''

# ✨ Style
sns.set_theme(style="whitegrid", palette="Set2")
plt.rcParams.update({
    'axes.titlesize': 16,
    'axes.labelsize': 14,
    'axes.titleweight': 'bold',
    'legend.fontsize': 12,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12
})

# ✅ Data cleaning
crop_df_filtered['value'] = pd.to_numeric(crop_df_filtered['value'], errors='coerce')

# ✅ Separate Yield and Production
yield_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'yield'].copy()
prod_df = crop_df_filtered[crop_df_filtered['element'].str.lower() == 'production'].copy()

# --- Detect outliers using IQR

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_df = df[(df[column] < lower_bound) | (df[column] > upper_bound)].copy()
    return outliers_df, lower_bound, upper_bound

yield_outliers, y_lb, y_ub = detect_outliers(yield_df, 'value')
prod_outliers, p_lb, p_ub = detect_outliers(prod_df, 'value')

# ✅ Flag outliers
yield_df['Outlier'] = yield_df['value'].apply(lambda x: 'Outlier' if (x < y_lb or x > y_ub) else 'Normal')
prod_df['Outlier'] = prod_df['value'].apply(lambda x: 'Outlier' if (x < p_lb or x > p_ub) else 'Normal')

# --- Visualization

fig, axes = plt.subplots(1, 2, figsize=(20, 7))

# ▶️ Yield plot
sns.boxplot(y='value', data=yield_df, ax=axes[0], showfliers=False, color='lightgray')
sns.stripplot(y='value', data=yield_df, hue='Outlier', dodge=False, ax=axes[0], alpha=0.5, jitter=True, size=5)
axes[0].set_title("Yield Distribution with Outliers Highlighted")
axes[0].set_ylabel("Yield (kg/ha)")
axes[0].legend(title="Status", loc="upper left",fontsize=14)

# ▶️ Production plot
sns.boxplot(y='value', data=prod_df, ax=axes[1], showfliers=False, color='lightgray')
sns.stripplot(y='value', data=prod_df, hue='Outlier', dodge=False, ax=axes[1], alpha=0.5, jitter=True, size=5)
axes[1].set_title("Production Distribution with Outliers Highlighted")
axes[1].set_ylabel("Production (tons)")
axes[1].legend(title="Status", loc="upper left",fontsize=14)

plt.suptitle("Outlier Detection in Yield and Production", fontsize=20, weight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# ✅ Print example rows
print("🔎 Yield Outliers (sample):")
print(yield_outliers[['area', 'item', 'year', 'value']].head(10))

print("\n🔎 Production Outliers (sample):")
print(prod_outliers[['area', 'item', 'year', 'value']].head(10))
```
